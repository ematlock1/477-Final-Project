```python
# import statements
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading in the Dataset
cardio_df = pd.read_csv('Cardiology - Sheet1.csv')
cardio_df
```

```python
# seperation of the datasets into two different ones based on the values of the NaN
valence_df = cardio_df[0:7]
print(valence_df)

obstacles_df = cardio_df[7:]
print(cardio_df)
```

```python
valence_df
```

```python
valence_df = valence_df.drop([0])
valence_df
```

```python
# plotting the chi-squared test results
colors = sns.color_palette("hls", 6)
sns.barplot(valence_df, x='Code', y='X^2 statistic', palette=colors, hue = 'Code', legend =False)
plt.xticks(rotation=90) 
plt.title("X^2 statistics for valences in IP and TC cardiology")
plt.show()
```

```python
colors = sns.color_palette("hls", 2)

#Looking at the differences in opinion between the perceptions of telemedicine 
valence_df['TC %'] = valence_df['TC %'].str.split().str[0]
valence_df['IP (%)'] = valence_df['IP (%)'].str.split().str[0]
valence_df

#putting the graphs onto on plot
fig, axes = plt.subplots(1,2, figsize=(12,6))

#perception of TC between two
valence_tc = valence_df[0:2]
valence_tc_pos = pd.DataFrame({'Frequency': [59, 13], 'Treatment Recieved': ['Telecardiology', 'In Person']})
sns.barplot(valence_tc_pos, x='Treatment Recieved', y='Frequency', palette=colors, ax=axes[0], hue = 'Treatment Recieved', legend =False)
axes[0].set_title("Positive Perception of Telecardiology after Recieving Treatment")

valence_tc_neg = pd.DataFrame({'Frequency': [0, 55], 'Treatment Recieved': ['Telecardiology', 'In Person']})
sns.barplot(valence_tc_neg, x='Treatment Recieved', y='Frequency', palette=colors, ax=axes[1], hue = 'Treatment Recieved', legend =False)
axes[1].set_title("Negative Perception of Telecardiology after Recieving Treatment")
axes[1].set_ylim(0, 60)

plt.tight_layout()
plt.show()
```

```python
#putting the graphs onto on plot
fig, axes = plt.subplots(1,2, figsize=(12,6))

#perception of TC between two
valence_ip = valence_df[2:4]
valence_ip_pos = pd.DataFrame({'Frequency': [1, 54], 'Treatment Recieved': ['Telecardiology', 'In Person']})
sns.barplot(valence_ip_pos, x='Treatment Recieved', y='Frequency', palette=colors, ax=axes[0], hue = 'Treatment Recieved', legend =False)
axes[0].set_title("Positive Perception of In Person after Recieving Treatment")
axes[0].set_ylim(0, 60)

valence_ip_neg = pd.DataFrame({'Frequency': [0, 2], 'Treatment Recieved': ['Telecardiology', 'In Person']})
sns.barplot(valence_ip_neg, x='Treatment Recieved', y='Frequency', palette=colors, ax=axes[1], hue = 'Treatment Recieved', legend =False)
axes[1].set_title("Negative Perception of In Person after Recieving Treatment")
axes[1].set_ylim(0, 60)

plt.tight_layout()
plt.show()
```
